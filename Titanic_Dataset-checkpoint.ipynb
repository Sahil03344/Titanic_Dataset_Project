```python
tit['Age'] = tit['Age'].fillna(tit['Age'].mean())
tit['Cabin'] =tit['Cabin'].bfill()
```

```python
tit.dropna(inplace = True)
```

```python
print("The Shape Of Dataset after dropping duplicate is ", tit.shape)
```

```python
Pc_count = tit['Pclass'].value_counts()
x = ["1st" , "2nd" , "3rd"]
y = [Pc_count[1] , Pc_count[2] , Pc_count[3]]
c = ["r" , "y" , "g"]

plt.xlabel("Passenger Class", fontsize = 12)
plt.ylabel("Passenger Count" , fontsize =12)
plt.title("PClass" , fontsize =15)
plt.bar(x,y, color =c , width =0.5)
plt.show()
```

```python
plt.xlabel("SibSp" , fontsize =12)
plt.ylabel("Parch" , fontsize =12)
size =[100]
plt.scatter(tit['SibSp'] , tit['Parch'] , s = size , c = "cyan" ,edgecolor ="darkred", alpha = 0.8)
plt.title("SibSp V/S Parch" , fontsize =15)
plt.show()
```

```python
correlation = tit.drop(['Name' , 'Sex' , 'Ticket' , 'Cabin' , 'Embarked'] , axis = 1).corr()
```

```python
sns.heatmap(correlation , annot = True , cmap = 'coolwarm')
plt.title("Features Correlation heatmap")
plt.show()
```
